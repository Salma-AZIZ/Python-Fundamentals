## **Instructions :**

### **1. Demandez les informations du client**
- **prenom** : Le pr√©nom du client.
- **nom** : Le nom du client.

### **2. G√©n√©rez un identifiant unique pour le client**
- **Identifiant** : Utilisez la premi√®re lettre du pr√©nom et les trois premi√®res lettres du nom pour g√©n√©rer un identifiant unique.

### **3. D√©clarez les informations d'un produit**
- **nom_produit** : Le nom du produit (cha√Æne de caract√®res).
- **prix_unitaire** : Le prix unitaire du produit (flottant).
- **quantite** : Le nombre d'unit√©s command√©es par le client (entier).


### **4. Calculez le montant total pour ce produit**
- Multipliez `prix_unitaire` par `quantite` pour obtenir le montant total avant remise.



### **5. Appliquez une remise**
- Si le montant total d√©passe 200 ‚Ç¨, une remise de 15% doit √™tre appliqu√©e. Utilisez une expression math√©matique.



### **6. Affichez un message personnalis√©**
- Le message doit inclure le nom complet du client, son identifiant, les d√©tails de la commande (produit et quantit√©), et le montant total √† payer apr√®s remise.



### **7. Calculez la somme totale √©conomis√©e**
- Si une remise est appliqu√©e, calculez et affichez le montant √©conomis√© par le client.

## **Exemple de Donn√©es :**

### **Exemple de Sortie Attendue :**

````python
# 1. Demandez les informations du client
prenom = input("Entrez le pr√©nom du client : ")
nom = input("Entrez le nom du client : ")

# 2. G√©n√©rez un identifiant unique pour le client
identifiant = prenom[0].upper() + nom[:3].lower()

# 3. D√©clarez les informations d'un produit
nom_produit = input("Entrez le nom du produit : ")
prix_unitaire = float(input("Entrez le prix unitaire du produit (‚Ç¨) : "))
quantite = int(input("Entrez le nombre d'unit√©s command√©es : "))

# 4. Calculez le montant total pour ce produit
montant_total = prix_unitaire * quantite

# 5. Appliquez une remise si applicable (sans utiliser if)
remise = 0.15 * montant_total * (montant_total > 200)  # La remise est de 15% si montant_total > 200, sinon 0
montant_a_payer = montant_total - remise

# 6. Affichez un message personnalis√©
print(f"\nBonjour, {prenom} {nom} (ID : {identifiant})")
print(f"Vous avez achet√© {quantite} {nom_produit}(s).")
print(f"Le montant total √† payer apr√®s remise est de : {montant_a_payer:.2f} ‚Ç¨.")

# 7. Calculez la somme totale √©conomis√©e et l'afficher (sans utiliser if)
print(f"Vous avez √©conomis√© : {remise:.2f} ‚Ç¨ gr√¢ce √† la remise.")
````
