### **B. Op√©rateurs de Comparaison**

Les **op√©rateurs de comparaison** permettent de comparer deux valeurs. Ils renvoient un r√©sultat bool√©en (`True` ou `False`).

| Op√©rateur | Description               | Exemple         |
|-----------|---------------------------|-----------------|
| `==`      | √âgal √†                    | `5 == 5` ‚Üí `True`  |
| `!=`      | Diff√©rent de              | `5 != 3` ‚Üí `True`  |
| `>`       | Plus grand que            | `7 > 3` ‚Üí `True`   |
| `<`       | Plus petit que            | `3 < 5` ‚Üí `True`   |
| `>=`      | Plus grand ou √©gal √†      | `5 >= 5` ‚Üí `True`  |
| `<=`      | Plus petit ou √©gal √†      | `3 <= 4` ‚Üí `True`  |

### **Pratiquez : Testez Vos Conditions !**

````python
# Votre code ici
x = 15
y = 20
print(x < y)  # R√©sultat attendu : True
````

### **C. Op√©rateurs Logiques**

Les **op√©rateurs logiques** sont utilis√©s pour combiner plusieurs conditions ensemble.

| Op√©rateur | Description               | Exemple                       |
|-----------|---------------------------|-------------------------------|

| `not`     | Inverse la valeur de la condition                  | `not (5 > 3)` ‚Üí `False`        |

## **6. Acc√©der aux √âl√©ments d'une Cha√Æne (Indexation et Slicing)**

Les cha√Ænes de caract√®res en Python sont index√©es √† partir de `0`. Vous pouvez aussi acc√©der √† une sous-partie d'une cha√Æne avec le **slicing** en utilisant `[start:end]`.
